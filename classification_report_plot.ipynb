```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
#crs_df = pd.read_csv("Resources/baseline_scores.csv")
#crs_df = pd.read_csv("Resources/PCA_analysis_scores.csv")
#crs_df = pd.read_csv("Resources/random_under_sampler_scores.csv")
#crs_df = pd.read_csv("Resources/standard_scalar_scores.csv")
#crs_df = pd.read_csv("Resources/removing_all_no_disease_records_scores.csv")
crs_df = pd.read_csv("Resources/clean_up_monkey_pox_negative_records_scores.csv")
crs_df.head()
```

```python
# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = crs_df.plot(kind="bar", figsize=(10,4))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(crs_df["Model"], rotation=0)

plt.title("Clean Up Monkey Pox Negative Records Scores")
plt.show()
plt.tight_layout()
```
